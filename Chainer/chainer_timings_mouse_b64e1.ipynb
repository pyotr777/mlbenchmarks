```python
class_weight = df[df["Parameter"].str.contains("class_weight")]
print class_weight["Value"].unique()
```

```python
df = df[df["Parameter"].str.contains("class_weight") == False]
df.iloc[:7]
```

```python
pat = re.compile("[0-9\.]+")
# Extract float number from a string
def parseFloat(s):
    global pat
    if type(s) is not str:
        return s
    f = pat.search(s)
    if f is not None:
        d = float(f.group())
        return d
    return None
```

```python
df[["Value"]]=df[["Value"]].applymap(parseFloat)
```

```python
grouped = df.groupby(by=["Address","Parameter"]).agg("sum")
grouped
```
